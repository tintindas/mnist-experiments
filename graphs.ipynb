```python
import json
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
file_1 = open("results/one_layer_nn_metrics.json")
metrics_1 = json.load(file_1)
file_2 = open("results/two_layer_nn_metrics.json")
metrics_2 = json.load(file_2)
```

```python
def convert_data(data):
    train_accuracy = data["train_accuracy"]
    train_loss = data["train_loss"]
    test_accuracy = data["test_accuracy"]
    test_loss = data["test_loss"]

    outer = []
    for values in train_accuracy.values():
        inner = list(values.values())
        outer.append(inner)
    train_accuracy = np.array(outer)

    outer = []
    for values in train_loss.values():
        inner = list(values.values())
        outer.append(inner)
    train_loss = np.array(outer)

    test_accuracy = np.array(list(test_accuracy.values()))

    test_loss = np.array(list(test_loss.values()))

    return {
        "train_accuracy": train_accuracy,
        "train_loss": train_loss,
        "test_accuracy": test_accuracy,
        "test_loss": test_loss
    }
```

```python
metrics_1 = convert_data(metrics_1)
metrics_2 = convert_data(metrics_2)
```

```python
train_accuracy = metrics_1["train_accuracy"]
train_accuracy_flat = train_accuracy.flatten()

# Noisy plot for batch wise accuracy
plt.plot(train_accuracy_flat)
plt.suptitle("Training Accuracy by Batch")
plt.ylabel("Accuracy")
plt.xlabel("Step")
```

```python
train_accuracy_avg = np.mean(train_accuracy, axis=1)

# Plot for epoch wise accuracy
plt.plot(train_accuracy_avg)
plt.suptitle("Training Accuracy by Epoch")
plt.ylabel("Accuracy")
plt.xlabel("Epoch")
plt.xticks(np.arange(1, 21))
```

```python
train_loss= metrics_1["train_loss"]
train_loss_flat = train_loss.flatten()

# Noisy plot for batch wise loss
plt.plot(train_loss_flat)
plt.suptitle("Training Loss by Batch")
plt.ylabel("Loss")
plt.xlabel("Step")
```

```python
train_loss_avg = np.mean(train_loss, axis=1)

# Plot for epoch wise accuracy
plt.plot(np.arange(1, 21), train_loss_avg)
plt.suptitle("Training Loss by Epoch")
plt.ylabel("Loss")
plt.xlabel("Epoch")
plt.xticks(np.arange(1, 21))
```

```python
test_accuracy = metrics_1["test_accuracy"]
test_loss = metrics_1["test_loss"]

plt.plot(np.arange(1, 21), test_accuracy)
plt.suptitle("Test Accuracy vs Epoch")
plt.ylabel("Accuracy")
plt.xlabel("Epoch")
plt.xticks(np.arange(1,21))
```

```python
plt.plot(np.arange(1, 21), test_loss)
plt.suptitle("Test Loss vs Epoch")
plt.ylabel("Loss")
plt.xlabel("Epoch")
plt.xticks(np.arange(1,21))
```

```python
plt.suptitle("Training vs. Epoch")
plt.plot(np.arange(1, 21), train_accuracy_avg, label="Training Accuracy")
plt.plot(np.arange(1, 21), test_accuracy, label="Test Accuracy")
plt.legend()
```

```python
plt.suptitle("Loss vs. Epoch")
plt.plot(np.arange(1, 21), train_loss_avg, label="Training Loss")
plt.plot(np.arange(1, 21), test_loss, label="Test Loss")
plt.legend()
```
